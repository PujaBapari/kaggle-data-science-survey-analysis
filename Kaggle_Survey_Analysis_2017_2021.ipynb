```python
import pandas as pd

# 1. Data load
df = pd.read_csv('kaggle_survey_2017_2021.csv')

# 2. First 5 row
display(df.head())

# 3. Dataset- row And column
print(f"\nDataset size: {df.shape}")

# 4. Column name
print(df.columns[:10])
```

```python
# 1. The first row (Questions) should be kept separate so that they can be referred to later if needed.
questions = df.iloc[0]

# 2. Dataset  1st row delete (Actual data)
df = df.drop(0).reset_index(drop=True)

# 3. Duplicate rows check
duplicate_count = df.duplicated().sum()
print(f"Total duplicate rows found: {duplicate_count}")

# 4. Duplicate rows delete kora
df = df.drop_duplicates()

# 5.Correct the Year column (where there is a '-')
df.rename(columns={'-': 'Year'}, inplace=True)
df['Year'] = df['Year'].astype(str).str.strip() # Extra space  remove

print("\nCleaning complete!")
print(f"Final dataset size: {df.shape}")
display(df.head(2))
```

```python
# 1. Creating a list of the columns we need
# Q1: Age, Q2: Gender, Q3: Country, Q4: Education, Q5: Job Title, Q6: Coding Experience
cols_to_keep = {
    'Year': 'Year',
    'Q1': 'Age',
    'Q2': 'Gender',
    'Q3': 'Country',
    'Q4': 'Education',
    'Q5': 'Job_Title',
    'Q6': 'Coding_Experience'
}

# 2. Create a new dataframe using only these columns
df_final = df[list(cols_to_keep.keys())].copy()
df_final.rename(columns=cols_to_keep, inplace=True)

# 4. Data summary
print("--- Selected Columns and First 5 Rows ---")
display(df_final.head())

# 5. Missing values check k
print("\n--- Missing Values in Each Column ---")
print(df_final.isnull().sum())
```

```python
# --- Handling Missing Values and Data Consistency ---

# 1. Fixing the 'Year' column using forward fill (ffill)
# This ensures that if any row is missing the year, it takes the previous valid year
if 'Year' in df.columns:
    df_final['Year'] = df_final['Year'].ffill()

# 2. Handling Missing Values for Categorical Columns
# Replacing NaN with 'Unknown' to maintain data integrity for visualization
categorical_cols = ['Education', 'Job_Title', 'Coding_Experience']
for col in categorical_cols:
    df_final[col] = df_final[col].fillna('Unknown')

# 3. Dropping rows where essential demographic info is missing
# Age, Gender, and Country are crucial for comparison; rows without these are less useful
df_final.dropna(subset=['Age', 'Gender', 'Country'], inplace=True)

# 4. Final verification of Missing Values
print("--- Missing Values Summary After Cleaning ---")
print(df_final.isnull().sum())

# 5. Reviewing the cleaned dataset size and structure
print(f"\nFinal dataset size after cleaning: {df_final.shape}")
print("\nFirst 5 rows of the processed data:")
display(df_final.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the visual style of the plots
sns.set(style="whitegrid")

# --- PART 1: General Demographic Dashboard (2x2 Grid) ---
plt.figure(figsize=(18, 14))

# 1. Top 10 Countries of Respondents
plt.subplot(2, 2, 1)
top_countries = df_final['Country'].value_counts().head(10)
sns.barplot(x=top_countries.values, y=top_countries.index, hue=top_countries.index, palette='viridis', legend=False)
plt.title('Top 10 Countries of Respondents', fontsize=14)
plt.xlabel('Count')

# 2. Gender Distribution (Top 3)
plt.subplot(2, 2, 2)
gender_data = df_final['Gender'].value_counts().head(3)
sns.countplot(data=df_final[df_final['Gender'].isin(gender_data.index)], x='Gender', hue='Gender', palette='coolwarm', legend=False)
plt.title('Gender Distribution (Top 3)', fontsize=14)

# 3. Top 8 Job Titles
plt.subplot(2, 2, 3)
top_jobs = df_final['Job_Title'].value_counts().head(8)
sns.barplot(x=top_jobs.values, y=top_jobs.index, hue=top_jobs.index, palette='magma', legend=False)
plt.title('Top 8 Job Titles', fontsize=14)
plt.xlabel('Count')

# 4. Number of Respondents per Year
plt.subplot(2, 2, 4)
sns.countplot(data=df_final, x='Year', hue='Year', palette='Set2', legend=False)
plt.title('Number of Respondents per Year', fontsize=14)

plt.tight_layout(pad=3.0)
plt.show()

# --- PART 2: Programming Language Trends (Comparison Over Years) ---
# Mapping the language columns from the original 'df'
languages_map = {
    'Q7_Part_1': 'Python',
    'Q7_Part_2': 'R',
    'Q7_Part_3': 'SQL',
    'Q7_Part_4': 'C',
    'Q7_Part_5': 'C++',
    'Q7_Part_6': 'Java'
}

lang_trends = []
for col, lang in languages_map.items():
    if col in df.columns:
        # Get count of people who selected the language for each year
        counts = df[df[col].notnull()].groupby('Year').size().reset_index(name='Count')
        counts['Language'] = lang
        lang_trends.append(counts)

# Combine and plot
df_lang_trends = pd.concat(lang_trends).sort_values('Year')

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_lang_trends, x='Year', y='Count', hue='Language', marker='o', linewidth=2.5)
plt.title('Programming Language Popularity Trends (2017-2021)', fontsize=15)
plt.ylabel('Number of Users')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
